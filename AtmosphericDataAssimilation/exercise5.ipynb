```python
hist,bin_edges = np.histogram(error)
hist, bin_edges = np.histogram(error, density=True)
plt.hist(error, bins='auto', normed=True)
muerror=error.mean()
stderror=error.std()
mu=muerror
sigma=stderror
x= np.linspace(mu-5*sigma, mu+5*sigma, 100)
plt.plot(x,mlab.normpdf(x, mu, sigma),'r')

plt.show() 
```
